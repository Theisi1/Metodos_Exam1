```python
import math


def aprox_log10_br(x):

    r = x - 1
    alpha = 1 / math.log(10)   
    return alpha * r



def ra_cons_redon(x, k):
  
    for _ in range(k):
        x = math.sqrt(x)
        x = float(f"{x:.6g}")   
    return x




def log10_bri(x, k=5):

    # x que reduce con las k raices cuadradas
    x_reducido = ra_cons_redon(x, k)

    # log10(1+r)
    log_aprox = aprox_log10_br(x_reducido)

    # log10(x) = 2^k * log10(x_reducido)
    return (2 ** k) * log_aprox


def main():
    numero = 3

    print("Brigs: 5 raíces, 6 cifras\n")
    
    resultado_briggs = log10_bri(numero, k=5)
    print(f"log10({numero}) con método de Briggs = {resultado_briggs}")

    # Valor real con función python
    valor_real = math.log10(numero)
    print(f"log10({numero}) real = {valor_real}")

    # Error absoluto
    error = abs(valor_real - resultado_briggs) / valor_real
    print(f"Error absoluto= {error}")


if __name__ == "__main__":
    main()
    
```
